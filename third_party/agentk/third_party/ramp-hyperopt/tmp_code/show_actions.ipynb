```python
import os 
import rampwf as rw
import ramphy as rh
from pathlib import Path
import shutil
import pickle as pkl

challenge = "kaggle_synthanic_v1_1_nX"
source_path = Path("/home/gpaolo/ramp-kits")
from pprint import pprint
```

```python
actions = [file for file in (source_path / challenge / 'actions').iterdir() if file.is_file()]
actions.sort()

loaded_actions = []
for action_path in actions:
    with open(action_path, 'rb') as f:
        loaded_actions.append(pkl.load(f))
```

```python
for action in loaded_actions[-10:]:
    print()
    print(action.name)
    print(action.start_time.strftime('%B %d, %Y, %H:%M:%S.%f'))
    pprint(action.kwargs)
```
